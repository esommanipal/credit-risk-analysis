```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/josephtoniodavid/Documents/credit-risk-analysis/data/cr_loan_clean.csv')
```

```python
#H9 The higher a loan amount the more likely it is to default
fig, ax = plt.subplots()
ax.figure.set_size_inches(6, 8)
sns.boxplot(x='loan_status', y='loan_amnt',  data=df, showfliers=False, palette=["b", "r"])
# set xticks labels
plt.xticks([0, 1], ['Non-Default', 'Default'])
# display plots
plt.show()
```

```python
df_default = df[df['loan_status'] == 1].copy()
```

```python
# H10 Young people have a higher chance of defaulting
fig, ax = plt.subplots()

# sets the figure size in inches
ax.figure.set_size_inches(25, 8)

# plots the graph
sns.countplot(x='person_age', hue='loan_status', data=df_default, palette=["r"])

# sets plot title
ax.set_title("Defaults count per age", fontsize=20)

# sets the legend
plt.legend(title="Loan status", labels=['Default'], title_fontsize=15, fontsize=15)

# display plots
plt.show()
```

```python
#H1 Loans with medical intent have a higher probability of default
med = df[df['loan_intent'] == 'MEDICAL']
m=med['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('MEDICAL')
print(m)
#med_def=med[med['']='Y']
debt=df[df['loan_intent'] == 'DEBTCONSOLIDATION']
d=debt['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('DEBTCONSOLIDATION')
print(d)
ed=df[df['loan_intent'] == 'EDUCATION'] 
e=ed['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('EDUCATION')
print(e)
home=df[df['loan_intent'] == 'HOMEIMPROVEMENT'] 
h=home['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('HOMEIMPROVEMENT')
print(h)
per=df[df['loan_intent'] == 'PERSONAL'] 
p=per['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('PERSONAL')
print(p)
ven=df[df['loan_intent'] == 'VENTURE'] 
v=ven['cb_person_default_on_file'].value_counts(normalize=True) * 100
print('VENTURE')
print(v)
#med[med['']='Y']
```

```python
#H6 People with higher loan percent income are more likely to default on loans as compared to people with lower loan percent income.
fig, ax = plt.subplots()
ax.figure.set_size_inches(6, 8)
sns.boxplot(x='loan_status', y='loan_percent_income',  data=df, showfliers=False, palette=["b", "r"])
# set xticks labels
plt.xticks([0, 1], ['Non-Default', 'Default'])
# display plots
plt.show()
```

```python
#H5 Homeowners are less likely to default on loans
# uses the variable ax for single a Axes
fig, ax = plt.subplots()

# sets the figure size in inches
ax.figure.set_size_inches(25, 8)

# plots the graph
sns.countplot(x='person_home_ownership', hue='loan_status', data=df_default, 
              order=df_default['person_home_ownership'].value_counts().index, palette=["r"])

# sets plot title
ax.set_title("Defaults count per home ownership", fontsize=20)

# sets the legend
plt.legend(title="Loan status", labels=['Default'], title_fontsize=15, fontsize=15)

# display plots
plt.show()
```
